```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
average_travel_times = pd.read_csv('data/average_travel_times.csv')
bus_stops = pd.read_csv('data/PATI_BUS_Stops.csv')
```

```python
unique_route_ids = set(average_travel_times['route_id'].unique())
print(unique_route_ids)
```

```python
def filter_fn(row):
    available_routes = set(row["Routes"].split(","))
    new = available_routes.intersection(unique_route_ids)
    return 0 if (len(new) == 0) else repr(new)    


print(bus_stops.size)
# filter out rows that contain the unique_route_ids
bus_stops['cleaned_routes'] = bus_stops.apply(filter_fn, axis=1)
bus_stops = bus_stops[bus_stops['cleaned_routes'] != 0]

# print the filtered bus stops
print(bus_stops.size)
```

```python
lst = []
for index, row in bus_stops.iterrows():
    routes = eval(row['cleaned_routes'])
    for route in routes:
        lst.append([route, row['Stop_ID'], row['Stop_Name'], row['X'],row['Y']])

route_stop = pd.DataFrame(lst,columns=['route_id','stop_id','stop_name','stop_lat','stop_lon'])
print(route_stop.head())
route_stop.to_csv('data/route_stop.csv', index=False)
```

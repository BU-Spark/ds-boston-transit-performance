```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('data/stops.csv')
df2=pd.read_csv("data/stops_to_neighborhoods.csv")
```

```python
print(df1.columns)
print(df2.columns)
```

```python
df2.isnull().sum()
```

```python
df2=df2.dropna()
```

```python
merged_data=pd.merge(df1,df2,on='stop_id')
```

```python
merged_data
```

```python
merged_data1=merged_data.copy()
merged_data1['route_id'] = merged_data1['route_id'].str.replace("'", '')
```

```python
merged_data1['route_id'] = merged_data1['route_id'].str.replace(' ', ',')
merged_data1
```

```python
merged_data1['route_id']=merged_data1['route_id'].astype(str)
```

```python
# Assuming 'route_id' is read as strings like '[191,SL4,SL5]'
merged_data1['route_id'] = merged_data1['route_id'].apply(lambda x: x.strip('[]').split(','))
merged_data1
```

```python
exploded_routes=merged_data1.explode('route_id')
exploded_routes['route_id']=exploded_routes['route_id'].astype(str)
```

```python
route_to_neigh=exploded_routes.groupby('route_id').agg({'stop_id': list, 'neighborhood': list}).reset_index()
route_to_neigh
```

```python
route_to_neigh.to_csv('data/routes_to_neighborhoods_new.csv',index=False)
```

```python
route_to_neigh1=exploded_routes.groupby('neighborhood').agg({'stop_id': list, 'route_id': list}).reset_index()
route_to_neigh1
```

```python
route_to_neigh1.to_csv('data/neighborhoods_to_routes_new.csv',index=False)
```

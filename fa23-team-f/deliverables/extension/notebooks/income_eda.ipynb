```python
import pandas as pd

# Load the dataset
file_path = '/Users/lijunyi/ds-boston-transit-performance/fa23-team-f/deliverables/extension/boston_income_by_zipcode.csv'
income_data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
income_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Descriptive statistics of the income column
income_stats = income_data['Income'].describe()

# Histogram of income distribution
plt.figure(figsize=(10, 6))
sns.histplot(income_data['Income'], kde=True)
plt.title('Distribution of Income across Boston ZIP Codes')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

income_stats
```

```python
# Grouping the data by neighborhood and calculating the average income for each neighborhood
neighborhood_income = income_data.groupby('Neighborhood')['Income'].mean().sort_values(ascending=False)

# Plotting the average income per neighborhood
plt.figure(figsize=(15, 8))
neighborhood_income.plot(kind='bar')
plt.title('Average Income by Neighborhood in Boston')
plt.xlabel('Neighborhood')
plt.ylabel('Average Income')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()

neighborhood_income.describe(), neighborhood_income.head(), neighborhood_income.tail()
```

Highest Average Income Neighborhoods: The neighborhoods with the highest average incomes are "West Broadway|South Boston" ($170,588), followed by "Beacon Hill" ($146,958), and "Financial District|Downtown" ($142,869).

Lowest Average Income Neighborhoods: The neighborhoods with the lowest average incomes include "Washington Park|Roxbury" ($31,900), "Franklin Field North|Roxbury" ($32,500), and "Mission Hill Projects|Jamaica Plain" ($38,300).

```python
import pandas as pd
```

```python
df_facilities = pd.read_csv('../data/MBTA_Website/MBTA_Systemwide_GTFS_Map.csv')
df_facilities.dropna(subset=['X', 'Y', 'wheelchair_facility'], inplace=True)
df_facilities.reset_index(drop=True, inplace=True)
df_facilities.head()
```

### Displaying the count of Bus Stops with Wheelchair facilities on the Map

```python
### Displaying Wheelchair facilities

import folium 
from folium.plugins import FastMarkerCluster

m = folium.Map(location=[42.3601, -71.0589], zoom_start=12) # centered at Boston
unique_stops = df_facilities.drop_duplicates(subset='stop_id', keep='first')
marker_cluster = FastMarkerCluster(data=list(zip(unique_stops['lat'], unique_stops['lon'])))

# Add each stop_id as a popup
for stop in unique_stops.itertuples():
    popup_html = f"Stop ID: {stop.stop_id}"
    folium.Marker([stop.lat, stop.lon], popup=folium.Popup(popup_html)).add_to(marker_cluster)

marker_cluster.add_to(m)
display(m)
```

```python
import folium 
from folium.plugins import HeatMap

# Assuming df_facilities contains columns 'stop_id', 'lat', 'lon'
m = folium.Map(location=[42.3601, -71.0589], zoom_start=12)  # centered at Boston
unique_stops = df_facilities.drop_duplicates(subset='stop_id', keep='first')

# Create a HeatMap layer based on the latitude and longitude
heat_data = list(zip(unique_stops['lat'], unique_stops['lon']))
HeatMap(heat_data).add_to(m)

# Display the map
display(m)
```

```python
import pandas as pd
from geopy.distance import geodesic

# Load the wheelchair accessibility dataset
df_accessibility = df_facilities  # Assuming unique_stops is your wheelchair accessibility dataset

# Load the neighborhood data
neighborhood_data = {
    'Place': ['Allston', 'Back Bay', 'Beacon Hill', 'Brighton', 'Charlestown', 'Chinatown', 'Dorchester', 'Downtown',
              'East Boston', 'Fenway', 'Hyde Park', 'Jamaica Plain', 'Longwood', 'Mattapan', 'Mission Hill', 'North End',
              'Roslindale', 'Roxbury', 'South Boston', 'South Boston Waterfront', 'South End', 'West End', 'West Roxbury'],
    'Latitude': [42.355537, 42.350707, 42.358708, 50.821463, 43.238348, 40.716491, 50.713206, 34.047996, 42.375097,
                 42.345187, 51.507489, 42.311605, 28.700850, 42.272321, 42.333265, 42.365097, 42.291209, 41.556828,
                 42.333431, 42.351938, 42.341310, 42.363919, 42.279265],
    'Longitude': [-71.132749, -71.079730, -71.067829, -0.140056, -72.424713, -73.996250, -2.437123, -118.252100, -71.039217,
                  -71.104599, -0.162237, -71.114384, -81.346558, -71.086995, -71.102029, -71.054495, -71.124497, -73.308892,
                  -71.049495, -71.049883, -71.077230, -71.063899, -71.149497]
}

df_neighborhoods = pd.DataFrame(neighborhood_data)

# Function to calculate the distance between two points using Vincenty formula
def haversine(coord1, coord2):
    return geodesic(coord1, coord2).miles

# Function to find the closest neighborhood to a facility
def find_neighborhood(facility_coord):
    distances = df_neighborhoods.apply(lambda row: haversine(facility_coord, (row['Latitude'], row['Longitude'])), axis=1)
    closest_neighborhood = df_neighborhoods.loc[distances.idxmin(), 'Place']
    return closest_neighborhood

# Assign neighborhood to each row in the wheelchair accessibility dataset
df_accessibility['Neighborhood'] = df_accessibility.apply(lambda row: find_neighborhood((row['lat'], row['lon'])), axis=1)

# Count the number of wheelchair-accessible facilities in each neighborhood
neighborhood_counts = df_accessibility['Neighborhood'].value_counts()

sorted_neighborhoods = neighborhood_counts.sort_values(ascending=False)

# Display the sorted neighborhoods
print("Neighborhoods sorted based on wheelchair accessibility:")
print(sorted_neighborhoods)
```

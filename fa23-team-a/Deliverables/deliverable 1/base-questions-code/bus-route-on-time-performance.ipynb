```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/fcz/BU_Google_Drive/BU_Courses/CS506/cs506-lab-project/MBTA_Reliability.csv")
df
```

```python
df["gtfs_route_desc"].unique()
df["mode_type"].unique()
```

```python
#keep only bus related rows
# df = df[~df["gtfs_route_desc"].isin(['Rapid Transit','Commuter Rail','Limited Service'])]
df = df[df["mode_type"]=="Bus"]
df
```

```python
df.shape
```

```python
# check metric type
df["metric_type"].unique()
```

This means that everything in this dataframe is on-time performance of bus. 

```python
df = df.dropna(subset=['otp_denominator', 'otp_numerator','cancelled_numerator'])
df.shape
```

No missing values found in 'otp_denominator', 'otp_numerator','cancelled_numerator'.

```python
# group gtfs_route_id to see on_time_rate
df['ot_rate'] = df['otp_numerator']/df['otp_denominator']
```

```python
type(df['gtfs_route_id'])
```

```python
grouped_route = df.groupby('gtfs_route_id')
grouped_rate = grouped_route['ot_rate'].mean().reset_index()
grouped_rate
```

```python
# plot bar chart in descending order

rate_sorted = grouped_rate.sort_values(by='ot_rate', ascending=False)
plt.figure(figsize=(8, 6))
plt.bar(x = rate_sorted['gtfs_route_id'], height=rate_sorted['ot_rate'])
plt.title('Bar Chart in Descending Order')
plt.xlabel('Bus Routes')
plt.ylabel('On-time Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
# keep top 10  bus routes with the highest on-time rate
n = 10

top_routes = rate_sorted.head(n)


```

```python
# keep bottom 10  bus routes with the lowest on-time rate

bottom_routes = rate_sorted.tail(n)
bottom_routes
```

```python
plt.figure(figsize=(8, 6))
df.boxplot(column='ot_rate')
plt.title('Boxplot by gtfs_route_id')
plt.suptitle('')  # Remove the default title
plt.xlabel('Bus Routes')
plt.ylabel('On-time rates')
plt.show()
```
